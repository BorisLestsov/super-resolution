```python
import torch
import torch.nn as nn
from numpy.random import normal
from numpy.linalg import svd
from math import sqrt


def _get_orthogonal_init_weights(weights):
    fan_out = weights.size(0)
    fan_in = weights.size(1) * weights.size(2) * weights.size(3)

    u, _, v = svd(normal(0.0, 1.0, (fan_out, fan_in)), full_matrices=False)

    if u.shape == (fan_out, fan_in):
        return torch.Tensor(u.reshape(weights.size()))
    else:
        return torch.Tensor(v.reshape(weights.size()))


class Net(nn.Module):
    def __init__(self, upscale_factor):
        super(Net, self).__init__()

        self.relu = nn.ReLU()
        self.conv1 = nn.Conv2d(1, 64, (5, 5), (1, 1), (2, 2))
        self.conv2 = nn.Conv2d(64, 64, (3, 3), (1, 1), (1, 1))
        self.conv3 = nn.Conv2d(64, 32, (3, 3), (1, 1), (1, 1))
        self.conv4 = nn.Conv2d(32, upscale_factor ** 2, (3, 3), (1, 1), (1, 1))
        self.pixel_shuffle = nn.PixelShuffle(upscale_factor)

        self._initialize_weights()

    def forward(self, x):
        
        print(x.size())
        x = self.relu(self.conv1(x))
        x = self.relu(self.conv2(x))
        x = self.relu(self.conv3(x))
        x = self.pixel_shuffle(self.conv4(x))
        return x

    def _initialize_weights(self):
        self.conv1.weight.data.copy_(_get_orthogonal_init_weights(self.conv1.weight) * sqrt(2))
        self.conv2.weight.data.copy_(_get_orthogonal_init_weights(self.conv2.weight) * sqrt(2))
        self.conv3.weight.data.copy_(_get_orthogonal_init_weights(self.conv3.weight) * sqrt(2))
        self.conv4.weight.data.copy_(_get_orthogonal_init_weights(self.conv4.weight))
        
    def criterion(self, inp, targ, loss_function):
        pred = self.forward(inp)
        
        num_images = inp.size(0)
        res = 0.0
        
        for scale_index in range(len(targ)):
            for img_index in range(len(targ)):
                
                img_targ = targ[img_index][scale_index]
#                 print(targ, '<- img targ')
                img_pred = pred[scale_index][img_index, :, :, :]
                
                res += loss_function(img_targ, img_pred)
                
        return res / num_images
```

```python
import cv2

def MSE(image, target):
    
    #print(image.size(), target.size(), '<- input, target sizes')
    res = (image - target) ** 2
    n_elements = res.numel()
    
    return res.sum() / n_elements

def pseudohuber(image, target, eps=1.0e-6):
    res = torch.sqrt((image - target) ** 2 + eps)
    n_elements = res.numel()
    
    return res.sum() / n_elements


def get_inp_targ(path, num_scales=3, fraction=8):
    img = cv2.imread(path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    x_cut = img.shape[0] % fraction
    y_cut = img.shape[1] % fraction
    
    img = img[x_cut:, y_cut:, :].astype(float)
    img = img / 256.0
    
#     print(img.max())
    
#     imshow(img)
    targs = [img]
    
    scale_factor = 1.0
    for index in range(num_scales):
        scale_factor /= 2.0
        targ = cv2.resize(img, None, 
                          fx=scale_factor, fy=scale_factor, 
                          interpolation=cv2.INTER_CUBIC)
        targ = numpy.clip(targ, 0.0, 1.0)
        targs.append(targ)
        
    inp = targs[-1]
    targs = targs[:-1][::-1]
    
#     print(inp.max())
#     imshow(inp)
    
    inp = inp.swapaxes(1, 2).swapaxes(0, 1)
    inp = inp.reshape((1,) + inp.shape)
    inp = torch.Tensor(inp).cuda()
    inp = torch.autograd.Variable(inp)
    
    for targ_index in range(len(targs)):
        targs[targ_index] = torch.autograd.Variable(
            torch.Tensor(
                targs[targ_index].swapaxes(1, 2).swapaxes(0, 1)).cuda())
    
    return inp, targs
```

```python
import skimage.io
img_path = './train2014/COCO_train2014_000000000009.jpg'

inp, targ = get_inp_targ(img_path)

loss = generator.criterion(inp, [targ], pseudohuber)
loss.backward()

print(loss)

print(inp[0].size())

# imshow()
print(inp[0])
# imshow(inp[0].cpu().data.numpy().swapaxes(1, 0).swapaxes(2, 1))
```

```python
optimizer = torch.optim.Adam(
    generator.parameters(),
    lr=1.0e1)
```

```python
import os

dir_name = './train2014/'
images = os.listdir(dir_name)

step_index = 0

print(images[0])

for epoch_index in range(1000):
    for image_name in images:
        img_path = dir_name + image_name
        inp, targ = get_inp_targ(img_path)
        
        generator.zero_grad()
        loss = generator.criterion(inp, [targ], MSE)
        loss.backward()
        
        optimizer.step()
        step_index += 1
        
        if step_index % 1 == 0:
            print(loss.data[0], end=' ')
```
